```python
# plot ACP
fig5 = plt.figure(figsize = (8,8))
ax = fig5.add_subplot(1,1,1) 
ax.set_xlabel('Principal Component 1', fontsize = 15)
ax.set_ylabel('Principal Component 2', fontsize = 15)

ax.set_title('2 component PCA', fontsize = 20)
targets = [True,False]
colors = ['r', 'g']
for target, color in zip(targets,colors):
    indicesToKeep = finalDf['is_genuine'] == target
    ax.scatter(finalDf.loc[indicesToKeep, 'principal component 1']
               , finalDf.loc[indicesToKeep, 'principal component 2']
               , c = color
               , s = 50)
ax.legend(targets)
ax.grid()
fig5.savefig('2_component_PCA_plot')
```

###### Qualité de représentation – Les COS² (cosinus carré)

Pour calculer la qualité de représentation des individus sur les axes, nous devons d’abord calculer les carrés des distances à l’origine des individus, qui correspondent également à leur contribution dans l’inertie total

```python
#contribution des individus dans l'inertie totale
di = np.sum(x**2,axis=1)
df_di =pd.DataFrame({'ID':df.index,'d_i':di})
df_di.sort_values('d_i', ascending =False).head()
```

Concrètement, le billet avec l'id 0 et le billet avec l'id 166 sont les deux billets qui se démarquent
le plus des autres, et on les retrouve aux deux extrémités le premier est un vrai billet et le deuxième est un faux billet.

```python
#qualité de représentation des individus - COS2
cos2 = coord**2
for j in range(p):
    cos2[:,j] = cos2[:,j]/di
df_ri =pd.DataFrame({'id':df.index,'COS2_1':cos2[:,0],'COS2_2':cos2[:,1]})
df_ri.head()
```

```python
#vérifions la théorie - somme en ligne des cos2 = 1
print(np.sum(cos2,axis=1))
```

Conformément à la théorie, pour chaque individu, la somme des COS² sur l’ensemble des
facteurs est égale à 1

#### Contribution des individus aux axes (CTR)
Elles permettent de déterminer les
individus qui pèsent le plus dans la définition de chaque facteur

```python
#contributions aux axes
ctr = coord**2
for j in range(p):
    ctr[:,j] = ctr[:,j]/(n*eigval[j])

df_ctr =pd.DataFrame({'id':df.index,'CTR_1':ctr[:,0],'CTR_2':ctr[:,1]})
df_ctr.sort_values('CTR_1', ascending =False).head()
```

```python
from sklearn.metrics import classification_report, confusion_matrix
import itertools
def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
  
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print(" Matrix de confusion normalisée")
    else:
        print('Matrix de confusion sans normalisation')

    print(cm)

    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('Vrai label')
    plt.xlabel('Prediction label')
print(confusion_matrix(y_test, yhat, labels=[1,0]))
```

```python
# Compute confusion matrix
cnf_matrix = confusion_matrix(y_test, yhat, labels=[1,0])
np.set_printoptions(precision=2)


# Plot non-normalized confusion matrix
plt.figure()
plot_confusion_matrix(cnf_matrix, classes=['is_genuine= True','is_genuine=False'],normalize= False,  title='Matrix de confusion')
plt.savefig('Confusion_matrix')
```

#### Algorithme pour detecter des faux billets

```python

def prediction(data):
    result = data[['diagonal', 'height_left', 'height_right', 'margin_low', 'margin_up','length']]
    result = preprocessing.StandardScaler().fit_transform(result)
    result = pd.DataFrame(LR.predict_proba(result))
    result.columns = ['Faux','Vrai']
    result = pd.concat([data.id,round(result.Vrai,1),round(result.Faux,1)],axis=1)
    result.set_index('id',inplace=True)
    del result.index.name
    return result

```

Si cette probabilité est supérieure ou égale à 0.5, le billet sera considéré comme vrai. Dans le cas contraire, il sera considéré comme faux.
